```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
for ii in data.DESCR.split('\n'):
    print(ii)
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=[10,5])
sns.displot(df,
            x='petal width (cm)', hue="target",kind="kde", height=6,
            multiple="fill", clip=(0, None),
            palette="ch:rot=-.25,hue=1,light=.75",)
```

```python
plt.figure(figsize=[10,5])
sns.displot(df,
            x='petal length (cm)', hue="target",kind="kde", height=6,
            multiple="fill", clip=(0, None),
            palette="ch:rot=-.25,hue=1,light=.75",)
```

```python
plt.figure(figsize=[10,5])
sns.displot(df,
            x='sepal width (cm)', hue="target",kind="kde", height=6,
            multiple="fill", clip=(0, None),
            palette="ch:rot=-.25,hue=1,light=.75",)
```

```python
plt.figure(figsize=[10,5])
sns.displot(df,
            x='sepal length (cm)', hue="target",kind="kde", height=6,
            multiple="fill", clip=(0, None),
            palette="ch:rot=-.25,hue=1,light=.75",)
```

```python
plt.figure(figsize=[10,8])
sns.violinplot(data=df[df.columns[:-1]], palette="light:g", inner="points", orient="h")
```

```python
df.boxplot(rot=45)
```

```python
nBins = 10
fig,axs = plt.subplots(2,2,figsize=(10,10))
# fig.figure()
axs[0,0].hist(df[df.columns[0]],bins=nBins)
axs[0,0].set_title(df.columns[0])
axs[0,1].hist(df[df.columns[1]],bins=nBins)
axs[0,1].set_title(df.columns[1])
axs[1,0].hist(df[df.columns[2]],bins=nBins)
axs[1,0].set_title(df.columns[2])
axs[1,1].hist(df[df.columns[3]],bins=nBins)
axs[1,1].set_title(df.columns[3])
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=[10,10])

sns.boxplot(x=df.columns[-1], y=df.columns[0], data=df, ax=axs[0,0])
axs[0,0].set_xticks(range(3))
axs[0,0].set_xticklabels(data.target_names.tolist(), rotation=45)

sns.boxplot(x=df.columns[-1], y=df.columns[1], data=df, ax=axs[0,1])
axs[0,1].set_xticks(range(3))
axs[0,1].set_xticklabels(data.target_names.tolist(), rotation=45)

sns.boxplot(x=df.columns[-1], y=df.columns[2], data=df, ax=axs[1,0])
axs[1,0].set_xticks(range(3))
axs[1,0].set_xticklabels(data.target_names.tolist(), rotation=45)

sns.boxplot(x=df.columns[-1], y=df.columns[3], data=df, ax=axs[1,1])
axs[1,1].set_xticks(range(3))
axs[1,1].set_xticklabels(data.target_names.tolist(), rotation=45)

fig.tight_layout(pad=1.0);
plt.show()
```

```python
fig, axs = plt.subplots(2,2,figsize=[10,10])
sns.violinplot(data=df,x=df.columns[-1],y=df.columns[0],split=True,ax=axs[0,0])
axs[0,0].set_xticks(range(3))
axs[0,0].set_xticklabels(data.target_names.tolist(), rotation=45)

sns.violinplot(x=df.columns[-1], y=df.columns[1], data=df, ax=axs[0,1])
axs[0,1].set_xticks(range(3))
axs[0,1].set_xticklabels(data.target_names.tolist(), rotation=45)

sns.violinplot(x=df.columns[-1], y=df.columns[2], data=df, ax=axs[1,0])
axs[1,0].set_xticks(range(3))
axs[1,0].set_xticklabels(data.target_names.tolist(), rotation=45)

sns.violinplot(x=df.columns[-1], y=df.columns[3], data=df, ax=axs[1,1])
axs[1,1].set_xticks(range(3))
axs[1,1].set_xticklabels(data.target_names.tolist(), rotation=45)

fig.tight_layout(pad=1.0);
plt.show()
```

```python
sns.pairplot(df, hue="target", height = 2, palette = 'colorblind');
```

```python
corr = df.corr()
mask = np.triu(np.ones_like(corr,dtype=bool))
cmap = sns.diverging_palette(10, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
sns.heatmap(corr,mask=mask,annot=True)
```

```python
sns.clustermap(df.corr(), center=0, cmap="vlag",
                   dendrogram_ratio=(.1, .2),
                   cbar_pos=(.02, .32, .03, .2),
                   linewidths=.75, figsize=(12, 13))
```
